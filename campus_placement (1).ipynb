# 2. Univariate Analysis

```python
plt.figure(figsize = (15, 7))
plt.style.use('seaborn-white')

#Specialisation
plt.subplot(234)
ax=sns.countplot(x="specialisation", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("magma", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)

#Work experience
plt.subplot(235)
ax=sns.countplot(x="workex", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("cividis", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)

#Degree type
plt.subplot(233)
ax=sns.countplot(x="degree_t", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("viridis", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12,rotation=20)

#Gender
plt.subplot(231)
ax=sns.countplot(x="gender", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("hot", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)

#Higher secondary specialisation
plt.subplot(232)
ax=sns.countplot(x="hsc_s", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("rocket", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)

#Status of recruitment
plt.subplot(236)
ax=sns.countplot(x="status", data=cr, facecolor=(0, 0, 0, 0),
                 linewidth=5,edgecolor=sns.color_palette("copper", 3))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)
```

### conclusion from plots
* We have more male candidates than female
* We have candidates who did commerce as their hsc course and as well as undergrad
* Science background candidates are the second highest in both the cases
* Candidates from Marketing and Finance dual specialization are high
* Most of our candidates from our dataset don't have any work experience
* Most of our candidates from our dataset got placed in a company

```python
# box plot for salary distribution for placed students
sns.set(rc={'figure.figsize':(12,8)})
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True)
salry=cr[cr.salary !=0]
sns.boxplot(salry["salary"],ax=ax_box)
sns.distplot(salry["salary"],ax=ax_hist)
plt.show()
```

### Conclusion of salary distribution of placed students
* Many candidates who got placed received package between 2L-4L PA
* Only one candidate got around 10L PA
* The average of the salary is a little more than 2LPA

### distribution of all the percentage i.e. 10th,12th,degree,employability,mba

```python
plt.figure(figsize = (15, 7))
plt.style.use('seaborn-white')
plt.subplot(231)
sns.distplot(cr['ssc_p'])
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.subplot(232)
sns.distplot(cr['hsc_p'])
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.subplot(233)
sns.distplot(cr['degree_p'])
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.subplot(234)
sns.distplot(cr['etest_p'])
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.subplot(235)
sns.distplot(cr['mba_p'])
fig = plt.gcf()
fig.set_size_inches(10,10)
```

#### conclusion for the distribution of percentage:

# 3. Bivariate Analysis

```python
#converting the status of placed and not placed into 1 and 2 respectively
cr.status=cr.status.apply(lambda x:str(x).replace("Not Placed","2"))
cr.status=cr.status.apply(lambda x:str(x).replace("Placed","1"))
#cr.status.unique()
cr.status=pd.to_numeric(cr.status,errors='coerce')
cr.status.unique()
```

### employability score vs salary

```python
#sns.set(rc={'figure.figsize':(12,8)})
sns.set(style="white", color_codes=True)
sns.jointplot(x=cr["etest_p"], y=cr["salary"], kind='kde', color="skyblue")
```

#### Conclusion of joint plot for employability vs salary
* Most of the candidates scored around 60 percentage got a decent package of around 3 lakhs PA
* Not many candidates received salary more than 4 lakhs PA


### 10th score vs salary

```python
sns.set(style="white", color_codes=True)
sns.jointplot(x=salry["ssc_p"], y=salry["salary"], kind='reg', color="skyblue")
```

#### Conclusion of joint plot for 10th percentage vs salary
* Most of the candidates scored around 70-80 percentage got a decent package of around 3 lakhs PA
* Not many candidates received salary more than 5 lakhs PA

### 12th score vs salary

```python
sns.set(style="white", color_codes=True)
sns.jointplot(x=salry["hsc_p"], y=salry["salary"], kind='reg', color="red")
```

#### Conclusion of joint plot for 12th % vs salary
* Most of the candidates scored around 60-70 percentage got a decent package of around 3 lakhs PA
* Not many candidates received salary more than 4 lakhs PA

### MBA score vs Salary

```python
sns.set(style="white", color_codes=True)
sns.jointplot(x=salry["mba_p"], y=salry["salary"], kind='reg', color="skyblue")
```

#### Conclusion of joint plot for MBA % vs salary
* Most of the candidates scored around 55-65 percentage got a decent package of around 2-3 lakhs PA
* Not many candidates received salary more than 6 lakhs PA
* Less number of candidates received salary between 4-6 lakhs PA

### Degree vs Placement_staus

```python
comm_placed = cr[cr.degree_t=="Comm&Mgmt"]
Sci_placed = cr[cr.degree_t=="Sci&Tech"]
Others_placed = cr[cr.degree_t=="Others"]
plt.figure(figsize=(4,4))
sns.countplot(x='status',data=comm_placed,palette='mako')
plt.figure(figsize=(4,4))
sns.countplot(x='status',data=Sci_placed,palette='mako')
plt.figure(figsize=(4,4))
sns.countplot(x='status',data=Others_placed,palette='mako')
```

#### Observation of count plot for Degree vs Placement_status
* Almost 70% of Commerce Students are Placed
* Almost 67% of Science Students are Placed
* Almost 55% of Students who neither opted science nor opted commerce are not placed

#### Conclusion of count plot for Degree vs Placement_status
* Commerce is the most demanded degree by placement company
* Degree other than commerce and science is less demanded by placement company

### Work_Experience vs Placement_staus


```python
placed = cr[cr.status==1]
plt.figure(figsize=(5,5))
sns.countplot(x='status',hue='workex',data=placed,palette='mako')
```

#### Conclusion of count plot for Work_Experience vs Placement_staus
* Work Experice does not matter for placement Company
